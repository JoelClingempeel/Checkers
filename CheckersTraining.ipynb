```python
import numpy as np
import pandas as pd
import torch
import torch.utils.data
```

```python
checkers_data = pd.read_csv('checkers_games_processed.csv')
```

```python
checkers_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
checkers_data['Winner'] = (checkers_data['Winner']+1)//2
```

```python
checkers_data
```

```python
Y_train = checkers_data['Winner']
X_train = checkers_data.drop('Winner', axis = 1)
                     
Y_train = Y_train.values
X_train = X_train.values
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.layer1 = nn.Linear(32, 32)
        self.layer2 = nn.Linear(32, 2)
    
    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = self.layer2(x)
        return x
        
net = Net()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(X_train, Y_train, test_size = 0.1)
```

```python
batch_size = 100
num_epochs = 4
learning_rate = 0.01
batch_no = len(x_train) // batch_size
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)
```

```python
from sklearn.utils import shuffle
from torch.autograd import Variable

test_var = Variable(torch.FloatTensor(x_val), requires_grad=True)

for epoch in range(num_epochs):
    if epoch % 5 == 0:
        print('Epoch {}'.format(epoch+1))
    x_train, y_train = shuffle(x_train, y_train)
    # Mini batch learning
    for i in range(batch_no):
        start = i * batch_size
        end = start + batch_size
        x_var = Variable(torch.FloatTensor(x_train[start:end]))
        y_var = Variable(torch.LongTensor(y_train[start:end]))
        # Forward + Backward + Optimize
        optimizer.zero_grad()
        ypred_var = net(x_var)
        loss =criterion(ypred_var, y_var)
        loss.backward()
        optimizer.step()
        if i%200 == 0:
            print(i)
        if i%1000 == 0:
            with torch.no_grad():
                result = net(test_var)
            values, labels = torch.max(result, 1)
            num_right = np.sum(labels.data.numpy() == y_val)
            print('Accuracy {:.2f}'.format(num_right / len(y_val)))
```

```python
# Evaluate the model on the training set for curiosity.
test_var = Variable(torch.FloatTensor(x_train), requires_grad=True)
with torch.no_grad():
    result = net(test_var)
values, labels = torch.max(result, 1)
num_right = np.sum(labels.data.numpy() == y_train)
print('Accuracy {:.2f}'.format(num_right / len(y_train)))
```
